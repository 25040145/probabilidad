```python
import itertools

# Lista de asistentes (pueden ser nombres)
asistentes = [f"A{i}" for i in range(1, 11)]

# Generar todas las combinaciones posibles de 4 asistentes
combinaciones = list(itertools.combinations(asistentes, 4))

# Mostrar cuántas combinaciones hay
print("Número total de combinaciones:", len(combinaciones))

# Mostrar todas las combinaciones
for c in combinaciones:
    print(c)
```
