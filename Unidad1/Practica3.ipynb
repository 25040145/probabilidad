```python
##creamos una lista vacia que representa la edad de 100 personas
import random
edades = []
##Hacemos un ciclo para generar 100 numeros aleatorios que representan la edad
for i in range(100):
  ##genereamos funcion randint y lo guardamos en variable numero_aleatorio
  numero_aleatorio =random.randint(18,70)
  #Agregamos el num aleatorio a la lista de valores
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
#Crear un DataFrame de pandas con los numeros aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
##Calcular la tabla de distribucion de frecuencias
tabla_frecuencias =df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad','Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')
##Imprimir la tabla de distribucion de frecuencias
print("Tabla de distribucion de frecuencias:")
print(tabla_frecuencias)
```

```python
##Tabla de Frecuencia Relativa
tabla_frecuencias['Frecuencia_Relativa'] = tabla_frecuencias['Frecuencia']/100
tabla_frecuencias
```

```python
##CALCULAR FRECUENCIA PORCENTUAL
tabla_frecuencias['Frecuencia_Porcentual'] = tabla_frecuencias['Frecuencia']/100*100
tabla_frecuencias
```

```python
##Calcular la frecuencia acumulada
tabla_frecuencias['Frecuencia_Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

Regla de sturges

1. Es una regla que sirve para cacular el numero de clases o intervalos en los que se debe de dividir un conjunto de datos
2. Establece que elnumero de clases es igual a uno mas en base dos del numero total de datos

```python
##Importamos la libreria math de funciones matematicas
import math
```

```python
##Volvemos a usar la lista de edades
print(edades)
```

```python
##Crear un DataFrame con las edades
df = pd.DataFrame({'Edad':edades})
df
```

```python
##Calcular el nuero de intervalos usando la regla de sturges
n = len(edades)
K = int(1+math.log2(n))
print("Valor de n =", n, "\n Numero de intervalos =",K)
```

```python
##Crear la tabla de frecuencia
tabla_frecuencia = pd.cut(df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns =['Clase', 'Frecuencia']
tabla_frecuencia
```

```python
##Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
##Generar columnas siguientes
tabla_frecuencia['Frecuencia Relativa'] = tabla_frecuencia['Frecuencia']/n
tabla_frecuencia['Frecuencia Porcentual'] =tabla_frecuencia['Frecuencia']/n*100
tabla_frecuencia['Frecuencia Acumulada'] =tabla_frecuencia['Frecuencia'].cumsum()
tabla_frecuencia['Frecuencia Relativa Acumulada'] =tabla_frecuencia['Frecuencia Relativa'].cumsum()
tabla_frecuencia['Frecuencia Porcentual Acumulada'] =tabla_frecuencia['Frecuencia Porcentual'].cumsum()
##Imprimir la tabla de frecuencia
print(tabla_frecuencia)
```
