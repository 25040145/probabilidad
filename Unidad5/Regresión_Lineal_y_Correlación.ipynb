## *Regresión Lineal y Correlación*

```python
from sklearn.linear_model import LinearRegression # Modelo de regresion Lineal
import numpy as np

# Datos de entrada
periodos = [2018, 2019, 2020, 2021, 2022, 2023]
inversion = np.array( [2, 3, 5, 4, 11, 5] ).reshape(-1, 1)
ganancias = [20, 25, 34, 30, 40, 31]

# Crear y entrenar el modelo de regresion Lineal
modelo = LinearRegression()
modelo.fit(inversion, ganancias)

# Predecir la ganancia para el año 2024
ganancia_2024 = modelo.predict( [[8]] )

print("La ganancia esperada para el año 2024 es de:", ganancia_2024[0])

# Coeficientes de la regresion
print("Coeficiente de la pendiente (m):", modelo.coef_[0])
print("Intercepcion de la recta (b):", modelo.intercept_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir tamaño de la figura
plt.figure(figsize=(6, 4))


# Datos de entrada
inversion = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]

# Dibujar el diagrama de dispersion
sns.scatterplot(x = inversion, y = ganancias)
plt.title("Diagrama de Dispersion de Inversion vs. Ganancias")
plt.xlabel("Inversion (millones)")
plt.ylabel("Ganancias (millones)")
plt.grid(True)
plt.show()
```

## *Regresión Lineal y Correlación - Agregar línea de regresión*

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Definir tamaño
plt.figure(figsize=(6, 4))

# Datos de entrada
inversion = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]

# Crear el diagrama de dispersion con linea de regresion
sns.set(style="whitegrid")
sns.regplot(x=inversion, y=ganancias, scatter=True, color= "blue")
plt.title("Diagrama de Dispersion con Linea de Regresion")
plt.xlabel("Inversion (millones)")
plt.ylabel("Ganancias (millones)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Datos de entrada
publicidad = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1)
ganancias = np.array([20, 25, 34, 30, 40, 31])

# Crear y entrenar el modelo de regresion Lineal
modelo = LinearRegression()
modelo.fit(publicidad, ganancias)

# Predecir la ganancias para una inversion de 8 en publicidad
inversion_prediccion = np.array([[8]])
ganancia_predicha = modelo.predict(inversion_prediccion)

print("La ganancia esperada para una inversion de 8 millones es:", ganancia_predicha[0])
print("Valor de la intercepcion", modelo.intercept_)
print("Valor de la pendiente", modelo.coef_)

# Crear el diagrama de dispersion con la linea de regresion
plt.figure(figsize=(6, 5))
plt.scatter(publicidad, ganancias, color="blue", label="Datos")
plt.plot(publicidad, modelo.predict(publicidad), color="red",label="Linea de Regresion")
plt.scatter(inversion_prediccion, ganancia_predicha, color="green", marker="x", s=100, label="Prediccion (8, {0:.2f})".format(ganancia_predicha[0]))

plt.title("Pronostico de ganancias a partir de la inversion en publicidad")
plt.xlabel("Inversion en Publicidad (millones)")
plt.ylabel("Ganancias Anuales (millones)")
plt.legend()
plt.grid(True)
plt.show()
```
